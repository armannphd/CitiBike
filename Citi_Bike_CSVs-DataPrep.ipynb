```python
import pandas as pd
import numpy as np
```

## New Analysis looking only at June 2021 - December 2021

```python
# create the dataframes for each CSV

df1 = pd.read_csv('06_2021.csv')
df2 = pd.read_csv('07_2021.csv')
df3 = pd.read_csv('08_2021.csv')
df4 = pd.read_csv('09_2021.csv')
df5 = pd.read_csv('10_2021.csv')
df6 = pd.read_csv('11_2021.csv')
df7 = pd.read_csv('12_2021.csv')
```

```python
# Create a list of all DataFrames
dataframes = [df1, df2, df3, df4, df5, df6, df7]

# Remove null values from each DataFrame
cleaned_dataframes = [df.dropna() for df in dataframes]

# Combine the cleaned DataFrames into one DataFrame
bike_half_2021 = pd.concat(cleaned_dataframes, ignore_index=True)

```

```python
# convert the date columns to datetime
bike_half_2021['started_at'] = pd.to_datetime(bike_half_2021['started_at'])
bike_half_2021['ended_at'] = pd.to_datetime(bike_half_2021['ended_at'])

```

```python
bike_half_2021.head()
```

```python
# Drop duplicates based on the 'Start Station ID' column and keep the first occurrence
unique_df = bike_half_2021.drop_duplicates(subset='start_station_name', keep='first')

unique_df.head()
```

```python
# Save the combined DataFrame to a CSV file
bike_half_2021.to_csv('bike_half_2021.csv', index=False)
```
